```python
# Initial imports
import os
import numpy as np
import pandas as pd
import seaborn as sns
import datetime
from pathlib import Path
```

```python
# Reading moneyball and salary data
file_path = Path("./Resources/Result/Moneyball.csv")
file_path_salary = Path("./Resources/2021 MLB Salary Combined.csv")

moneyball_df = pd.read_csv(file_path, index_col = 0)
salary_df = pd.read_csv(file_path_salary, index_col = 0)

moneyball_df.head(20)
```

```python
salary_df.head()
```

```python
salary_df = salary_df.drop(columns = ['Age','Status','Wavier Option', 'Base Salary', 'Signing Bonus', 'Adjusted Salary', 'Payroll %', 'Luxury Tax Salary', 'Injured'])
salary_df.head()
```

```python
combined_df = pd.concat([moneyball_df, salary_df], axis="columns", join="inner")
combined_df.head(20)
```

```python
expanded_moneyball_df = combined_df
expanded_moneyball_df['OPS Per Millionth Dollar'] = expanded_moneyball_df['Predicted'] / expanded_moneyball_df['Payroll Salary'] * 1000000
expanded_moneyball_df.to_csv(r'expanded moneyball.csv')
expanded_moneyball_df.head(20)
```

```python
expanded_moneyball_df['OPMD'] = expanded_moneyball_df['OPS Per Millionth Dollar']
expanded_moneyball_df = expanded_moneyball_df.drop(columns = 'OPS Per Millionth Dollar')
expanded_moneyball_df.reset_index(inplace=True)
expanded_moneyball_df = expanded_moneyball_df.sort_values(['Position', 'OPMD'], ascending=[True, False]).drop_duplicates(['Position']).reset_index(drop=True)
```
