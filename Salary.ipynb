```python
import os
import numpy as np
import pandas as pd
import seaborn as sns
import datetime
from pathlib import Path
import glob
```

```python
file_path = Path("Resources/Result/Moneyball.csv")
file_path_salary = Path("Resources/2021 MLB Salary Combined.csv")

moneyball_df = pd.read_csv(file_path, index_col = 0)
salary_df = pd.read_csv(file_path_salary, index_col = 0)

moneyball_df.head(20)
```

```python
salary_df = salary_df.drop(columns = ['Age','Status','Wavier Option', 'Base Salary', 'Signing Bonus', 'Adjusted Salary', 'Payroll %', 'Luxury Tax Salary', 'Injured'])
salary_df.head(20)
```

```python
combined_df = pd.concat([moneyball_df, salary_df], axis="columns", join="inner")
combined_df.head(20)
```

```python
combined_df.to_csv(r'Resources/Result/combined.csv')
```

```python
expanded_moneyball_df = combined_df
expanded_moneyball_df['OPS Per Millionth Dollar'] = expanded_moneyball_df['Predicted'] / expanded_moneyball_df['Payroll Salary'] * 1000000
expanded_moneyball_df.to_csv(r'Resources/Result/expanded moneyball.csv')
expanded_moneyball_df.head(20)
```

```python
# position_df = expanded_moneyball_df1.groupby('Position').max().drop(columns = ['Real', 'Predicted','Payroll Salary'])
sorted_df = expanded_moneyball_df.sort_values(by = ['Position', 'OPS Per Millionth Dollar'], ascending = False)
sorted_df['OPMD'] = sorted_df['OPS Per Millionth Dollar']
sorted_df = sorted_df.drop(columns = 'OPS Per Millionth Dollar')
sorted_df.reset_index(inplace=True)
sorted_df.set_index(['Position', 'OPMD'], inplace=True)
# sorted_df.set_index(['Position'], inplace=True)
sorted_df.head()
```

```python
from pandasql import sqldf

q="""SELECT DISTINCT Position, max([OPMD]) as OPMD FROM sorted_df GROUP BY Position"""
pysqldf = lambda q: sqldf(q, globals())
a_df = pysqldf(q)

a_df
```

```python
a_df.set_index(['Position', 'OPMD'], inplace=True)
```

```python
q="""SELECT s.* FROM sorted_df s INNER JOIN a_df a ON s.Position = a.Position AND s.OPMD = a.OPMD"""
pysqldf = lambda q: sqldf(q, globals())
a_df = pysqldf(q)

a_df.set_index('Position', inplace=True)
a_df
```

```python
a_df.to_csv(r'Resources/Result/final moneyball.csv')
```
